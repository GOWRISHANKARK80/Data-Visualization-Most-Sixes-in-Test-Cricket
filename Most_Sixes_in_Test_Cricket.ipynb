```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Most_Sixes_in_Test_Cricket.csv")
df
```

```python
pd.set_option("display.max_rows",None)
df
```

```python
import plotly.express as px
```

```python
px.pie(df,names="Player",values="6s",hole=0.4)
```

```python
px.sunburst(df,path=["Runs","Player"],values="Mat")
```

```python
# Assuming 'df' is your DataFrame with columns 'Player' and '6s'
top_5_players = df.sort_values(by='6s', ascending=False).head(5)
fig = px.bar(top_5_players,x='Player',y='6s',title='Top 5 Players with Most Sixes')
fig.show()
```

```python
fig=px.pie(df,names="Player",values="6s")
fig.update_traces(textposition="inside")
fig.show()
```
